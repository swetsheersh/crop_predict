```python
import numpy as np
cat_col=df_input.select_dtypes(include=object).columns.tolist()
```

```python
num_col=df_input.select_dtypes(include=np.number).columns.tolist()
```

```python
cat_col
```

```python
jovian.commit()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df_input=df_input.fillna('unknown')
```

```python
encoder=OneHotEncoder(sparse=False,handle_unknown="ignore")
```

```python
encoder.fit(df_input[cat_col])
```

```python
encoded_cols=list(encoder.get_feature_names(cat_col))
encoded_cols
```

```python
len(encoded_cols)
```

```python
df_input[encoded_cols]=encoder.transform(df_input[cat_col])
```

```python
df_input=df_input[num_col+encoded_cols]
```

```python
df_input
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(df_input[num_col])
```

```python
df_input.describe().loc[['min','max']]
```

```python
df_input[num_col]=scaler.transform(df_input[num_col])
```

```python
df_input
```

```python
df_input.describe().loc[['min','max']]
```

```python
jovian.commit()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df,val_df,train_target,val_target=train_test_split(df_input,df_target,test_size=0.2,random_state=42)
```

```python
train_df
```

```python
val_df
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 15)
```

```python
X_train = pd.DataFrame(pca.fit_transform(train_df))
X_test = pd.DataFrame(pca.transform(val_df))
```

```python
train_df.columns.tolist()
```

```python
from sklearn.svm import SVR
```

```python
classifier=SVR(kernel='rbf',shrinking=True)
```

```python
classifier.fit(X_train,train_target)
```

```python
predict_train=classifier.predict(X_train)
predict_train
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(train_target, predict_train)
```

```python
predict_test=classifier.predict(X_test)
predict_test
```

```python
mean_squared_error(val_target, predict_test)
```
